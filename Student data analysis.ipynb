categorizing language.grade column

We would cut the language.grade values and make 4 categories. Fail Good Great Excellent to describe it more using catigorize_col() function provided above

```python
def catigorize_col(df, col):

    edges = [1, 2, 3, 4, 5]
    labels = ['Fail', 'Good', 'Great', 'Excellent']

    df[col] = pd.cut(
        df[col],
        bins=edges,
        labels=labels,
        include_lowest=True
    )

    return df
```

```python
catigorize_col(df, 'language.grade')

print(df['language.grade'].unique())
```
